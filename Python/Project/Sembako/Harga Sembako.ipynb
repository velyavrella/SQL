### Data Visualization

```python
import matplotlib.pyplot as plt # visualisasi data
import seaborn as sns # visualisasi data
```

```python
data_df.plot.bar()
plt.title('Harga Sembago di Indonesia')
plt.legend(loc = 'upper left')
```

```python
prov = pd.Series(data_df.index)
sembako = data_df.iloc[:,0].sort_values()
```

```python
fig, ax = plt.subplots(figsize =(8, 10))
ax.barh(prov, sembako)

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Harga Beras di Indonesia',
             loc ='left')


plt.show()
```

```python
prov = pd.Series(data_df.index)
sembako = data_df.iloc[:,1].sort_values()

fig, ax = plt.subplots(figsize =(8, 10))
ax.barh(prov, sembako, color = 'green')

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Harga Minyak Goreng di Indonesia',
             loc ='left')


plt.show()
```
