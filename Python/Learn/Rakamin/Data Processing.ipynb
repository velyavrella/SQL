# Data Aggregation

### Melihat jumlah orang per provinsi

```python
df_dataset.groupby('province')['registration_id'].count()
```

### Melihat data dua pertama berdasarkan provinsi

```python
df_dataset.groupby(['province']).head(2)
```

### Tambahkan `.reset_index()` untuk membuat dataframe yang kolomnya rata

```python
df_dataset.groupby('province')['registration_id'].count().reset_index()
```

## Melihat rata-rata score tiap kelas

```python
df_dataset.groupby('class')['score'].mean().reset_index()
```

### Kemudian diurutkan dari rata-rata score tertinggi

```python
df_dataset.groupby('class')['score'].mean().reset_index().sort_values('score',ascending=False)
```

## Melihat score terbesar per provinsi per kelas

```python
df_dataset.groupby(['province','class'])['score'].max().reset_index()
```

### Usia termudanya untuk masing-masing kota dan provinsi

```python
df_dataset.groupby(['province','city'])['age'].min().reset_index()
```

## Rata-rata score per kota untuk orang-orang dari Jakarta

```python
df_dataset[df_dataset['province'] == 'Jakarta'].groupby('city')['score'].mean().reset_index()
```

## Banyak orang yang aktif, dan tidak aktif

```python
df_dataset.groupby('active')['registration_id'].count().reset_index()
```
