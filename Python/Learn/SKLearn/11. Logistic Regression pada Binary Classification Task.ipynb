## Dataset: SMS Spam Collection Data Set

```python
import pandas as pd

df = pd.read_csv('./SMSSpamCollection',
                sep = '\t', #separator tab
                header = None,
                names = ['label', 'sms'])

df.head()
```

```python
df['label'].value_counts()
```

* unbalanced datasets

## Training & Testing Dataset

```python
from sklearn.preprocessing import LabelBinarizer

X = df['sms'].values
y = df['label'].values

lb = LabelBinarizer()
y = lb.fit_transform(y).ravel() #multidimensi -> satu dimensi
lb.classes_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,
                                                   y,
                                                   test_size = 0.25,
                                                   random_state = 0)
print(X_train, '\n')
print(y_train)
```

## Feature Extraction dengan TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer=TfidfVectorizer(stop_words = 'english')

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print(X_train_tfidf)
```

## Binary Classification dengan Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
y_pred = model.predict(X_test_tfidf)

for pred, sms in zip(y_pred[:5], X_test[:5]):
    print(f'PRED : {pred} - SMS: {sms}\n')
```

## Evaluation Metrics pada Binary Classification
* Confusion Matrix
* Accuracy
* Precission & Recall
* F1 Score
* ROC

### Terminologi Dasar
* True positif
* True negatif
* False Positif
* False Negatif

## Confusion Matrix
Confusion Matrix seringkali juga dikenal sebagai error matrix

```python
from sklearn.metrics import confusion_matrix

matrix = confusion_matrix(y_test, y_pred)
matrix
```

```python
tn, fp, fn, tp = matrix.ravel()

print(f'TN: {tn}')
print(f'FP: {fp}')
print(f'FN: {fn}')
print(f'TP: {tp}')
```

```python
import matplotlib.pyplot as plt
plt.matshow(matrix)
plt.colorbar()

plt.title('Confusion Matrix')
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show
```

### Accuracy
Akurasi mengukur porsi dari hasil prediksi yang tepat dalam rentang 0-1.

dengan t: true, f: false
n: negatif, p:positif

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print(f'Accuracy : {acc}')
```

### Precission & Recall
Selain menggunakan accuracy, performa dari suatu classifier umumnya juga diukur berdasarkan nilai Precission and Recall. Keduanya dalam rentang 0-1.


### Precission or Positive Predictive Value (PPV)


```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)

print(f'Precission: {prec}')
```

### Recall or True Positive Rate (TPR) or Sensitivity


```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'Recall : {rec}')
```

### F1-Score
F1-Score atau F1-measure adalah harmonic mean dari precission dan recall.


```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1 : {f1}')
```

### ROC: Receiver Operating Characteristic
ROC menawarkan visualisasi terhadap perfoma dari classifier dengan membandingkan nilai Recall (TPR) dan nilai Fallout (FPR) dalam rentang 0-1.



```python
from sklearn.metrics import roc_curve, auc

# auc: area under the curve
# semakin kurva mendekati perfect classifier maka semakin besar pula nilai auc

prob_estimates = model.predict_proba(X_test_tfidf)

fpr, tpr, threshhold = roc_curve(y_test, prob_estimates[:, 1])
nilai_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, 'b', label = f'AUC = {nilai_auc}')
plt.plot([0,1], [0,1], 'r--', label = 'Random Classifier')

plt.title('ROC: Receiver Operating Characteristic')
plt.xlabel('Fallout of False Positive Rate')
plt.ylabel('Recall of True Positive Rate')
plt.legend()
plt.show()
```
