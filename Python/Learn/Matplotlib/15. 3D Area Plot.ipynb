```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Dataset

```python
n = np.linspace(-3.0, 3.0, 100)
    
x, y = np.meshgrid(n, n) # data area
z = np.sqrt(x**2 + y**2)
    
print(x.shape)
print(x.shape)
print(x.shape)
```

## 3D Wireframe

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_wireframe(x, y, z, color = 'orange')

ax.set(title = '3D Wireframe',
      xlabel = 'Sumbu X',
      ylabel = 'Sumbu Y',
      zlabel = 'Sumbu Z')

plt.show()
```

## 3D Surface Plot

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_surface(x, y, z, color = 'orange')
# ax.plot_surface(x, y, z, cmap = 'viridis')

ax.set(title = '3D Surface Plot',
      xlabel = 'Sumbu X',
      ylabel = 'Sumbu Y',
      zlabel = 'Sumbu Z')

plt.show()
```
