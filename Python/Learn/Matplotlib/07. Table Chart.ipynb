## Import modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Data

```python
data_uas = [ ['Bejo',70],
            ['Tejo', 83],
            ['Cecep', 82],
            ['Mati', 74],
            ['Karti', 71]
]
data_uas
```

## Table Plot dengan OO Style

```python
fig, ax = plt.subplots()

table = plt.table(cellText = data_uas,
                  loc = 'center')

table.set_fontsize(14)
table.scale(1, 4) # ukuran kolom, baris

ax.axis(False)

plt.show()
```

## Table PLot dengan pyplot Style

```python
table = plt.table(cellText = data_uas,
                 loc = 'center')

table.set_fontsize(14)
table.scale(1,4) # ukuran kolom, baris

ax = plt.gca()
ax.axis(False)

plt.show()
```
