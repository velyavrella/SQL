## Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Data

```python
mu, sigma = 100, 15 # nilai mean dan nilai standard deviation

x = mu + sigma * np.random.randn(1000) # normal distribution
x
```

```python
x.shape
```

## Histogram dengan pyplot Style

```python
plt.hist(x,
        bins = 50, # 50 kelompok data dg msg2 frekuensinya
        facecolor = 'g', # warna plot
        alpha = 0.75) # transparansi

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(50, 55, r'$\mu = 100,\ \sigma = 15$') # $ : format latex dan r : optional
plt.grid()

plt.show()
```

## Histogram dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x,
       bins = 50,
       facecolor = 'r',
       alpha = 0.75)

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(50, 55, r'$\mu =100,\ \sigma=15$')
ax.grid()

plt.show()
```
