## Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Simple Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data,
        labels = kategori,
        autopct = '%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')

plt.show()
```

## Pengaturan Legend dan Color

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e83364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
        labels = kategori,
        colors = warna,
        autopct = '%1.1f%%',
        startangle = 90)

plt.legend()
plt.title('Contoh Pie Plot')

plt.show()
```

## Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e83364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
        labels = kategori,
        colors = warna,
        explode = explode_var,
        autopct = '%1.1f%%',
        startangle = 90)

plt.title('Contoh Pie Plot')

plt.show()
```
