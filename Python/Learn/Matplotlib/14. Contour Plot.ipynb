## Import Modules

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Dataset

```python
n = np.linspace(-3.0, 3.0, 100)
    
x, y = np.meshgrid(n, n) # data area
z = np.sqrt(x**2 + y**2)
    
print(x.shape)
print(x.shape)
print(x.shape)
```

```python
x
```

```python
y
```

## Contour Plot

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

## Filled Contour Plot

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
