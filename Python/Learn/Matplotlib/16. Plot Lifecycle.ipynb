## Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Sample Dataset

```python
data = {'Item A' : 109438,
       'Item B' : 103569,
       'Item C' : 112214,
       'Item D' : 112591,
       'Item E' : 100934,
       'Item F' : 103660,
       'Item G' : 137351,
       'Item H' : 123381,
       'Item I' : 135841,
       'Item J' : 104437}
data
```

```python
items = tuple(data.keys())
items
```

```python
count = tuple(data.values())
count
```

## Simple Plot

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

## Pengaturan Style

```python
plt.style.available
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

## Pengaturan Tick Label

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
        rotation = 45,
        horizontalalignment = 'right')

plt.show()
```

## Pengaturan Format pada Ticker

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title = 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

## Pengaturan Label pada Sumbu (axis) dan Judul (title)

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title = 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

## Penambahan Garis (vertical/horizontal line) pada Plot

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
          ls = '--',
          color = 'r')

ax.set(title = 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

plt.show()
```

## Menyimpan Hasil Plot ke dalam suatu File

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
          ls = '--',
          color = 'r')

ax.set(title = 'Contoh Plot',
      xlabel = 'Jumlah',
      ylabel = 'Items')

fig.savefig('sales.png',
           transparent = False,
           dpi = 80)
```
